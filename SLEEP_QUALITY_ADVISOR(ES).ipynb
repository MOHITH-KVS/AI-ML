```python
import matplotlib.pyplot as plt
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from google.colab import files
import os

class SleepQualityAdvisor:
    def __init__(self):
        self.sleep_hours = 0
        self.stress_level = 0
        self.exercise_level = 0
        self.diet_quality = 0
        self.noise_level = 0
        self.caffeine_intake = 0
        self.screen_time = 0
        self.sleep_schedule_consistency = 0
        self.sleep_quality = ""

    def ask_questions(self):
        print("Welcome to the Enhanced Sleep Quality Advisor!")

        # Sleep Hours
        self.sleep_hours = self.ask_question_with_validation(
            "How many hours of sleep do you get on average?",
            ["1. Less than 6 hours", "2. 6-8 hours", "3. More than 8 hours"],
            {1: 5, 2: 7, 3: 9}
        )
        # Stress Level
        self.stress_level = self.ask_question_with_validation(
            "On a scale of 1-10, how stressed do you feel before bed?",
            ["1. 1-3 (Low Stress)", "2. 4-6 (Moderate Stress)", "3. 7-10 (High Stress)"],
            {1: 3, 2: 6, 3: 9}
        )
        # Exercise Level
        self.exercise_level = self.ask_question_with_validation(
            "How many days per week do you exercise?",
            ["1. 0-1 days", "2. 2-3 days", "3. 4 or more days"],
            {1: 1, 2: 3, 3: 5}
        )
        # Diet Quality
        self.diet_quality = self.ask_question_with_validation(
            "On a scale of 1-10, how healthy is your diet?",
            ["1. 1-3 (Poor Diet)", "2. 4-7 (Moderate Diet)", "3. 8-10 (Healthy Diet)"],
            {1: 3, 2: 6, 3: 9}
        )
        # Noise Level
        self.noise_level = self.ask_question_with_validation(
            "On a scale of 1-10, how quiet is your sleeping environment?",
            ["1. 1-3 (Noisy)", "2. 4-7 (Moderately Quiet)", "3. 8-10 (Very Quiet)"],
            {1: 3, 2: 6, 3: 9}
        )
         # Caffeine Intake
        self.caffeine_intake = self.ask_question_with_validation(
            "How many caffeinated drinks (coffee, tea, energy drinks) do you have per day?",
            ["1. 3 or more drinks", "2. 1-2 drinks", "3. None"],
            {1: 2, 2: 1, 3: 0}
        )
         # Screen Time
        self.screen_time = self.ask_question_with_validation(
            "How much screen time do you have in the hour before sleep?",
            ["1. More than 1 hour", "2. 30 minutes to 1 hour", "3. Less than 30 minutes"],
            {1: 2, 2: 1, 3: 0}
        )
        # Sleep Schedule Consistency
        self.sleep_schedule_consistency = self.ask_question_with_validation(
            "Do you maintain a consistent sleep-wake schedule?",
            ["1. No, my schedule varies", "2. Mostly, but weekends differ", "3. Yes, it is consistent"],
            {1: 0, 2: 1, 3: 2}
        )

    def ask_question_with_validation(self, question, options, valid_answers):
        print("\n" + question)
        for option in options:
            print(option)
        while True:
            try:
                choice = int(input("Enter the number corresponding to your choice: "))
                if choice in valid_answers:
                    return valid_answers[choice]
                else:
                    print("Invalid input. Please choose one of the provided options (1, 2, or 3).")
            except ValueError:
                print("Invalid input. Please enter a number corresponding to one of the options.")


    def evaluate_sleep_quality(self):
        score = 0

        # Sleep Hours
        score += 2 if 6 <= self.sleep_hours <= 8 else -1

        # Stress Level
        score += 2 if self.stress_level <= 3 else (-3 if self.stress_level >= 7 else -1)

        # Exercise Level
        score += 3 if self.exercise_level >= 4 else (1 if self.exercise_level >= 2 else -2)

        # Diet Quality
        score += 2 if self.diet_quality >= 8 else (-3 if self.diet_quality <= 4 else 1)

        # Noise Level
        score += 2 if self.noise_level >= 8 else (-2 if self.noise_level <= 3 else 1)

        # Caffeine Intake
        score -= 2 if self.caffeine_intake >= 2 else (-1 if self.caffeine_intake == 1 else 1)

        # Screen Time
        score -= 2 if self.screen_time >= 2 else (-1 if self.screen_time == 1 else 1)

        # Sleep Schedule Consistency
        score += 2 if self.sleep_schedule_consistency == 2 else (1 if self.sleep_schedule_consistency == 1 else -1)

        # Determine Sleep Quality
        self.sleep_quality = (
            "Poor" if score <= -5 else
            "Fair" if -5 < score <= 0 else
            "Good" if 0 < score <= 5 else
            "Excellent"
        )

    def provide_advice(self):
        if self.sleep_quality == "Poor":
            print("\nAdvice: You may need to make several lifestyle adjustments to improve your sleep quality.")

            print("\n**Sleep Duration and Routine**")
            print(" - Aim for at least 7-8 hours of sleep each night.")
            print(" - Set a regular bedtime and wake-up time, even on weekends.")
            print(" - Avoid long naps during the day, as they can disrupt nighttime sleep.")

            print("\n**Stress and Relaxation**")
            print(" - Try relaxation techniques like deep breathing, progressive muscle relaxation, or mindfulness meditation.")
            print(" - Consider natural remedies, such as drinking chamomile or valerian tea before bed to help calm your mind.")
            print(" - Journaling before bed can help clear your mind of stressful thoughts.")

            print("\n**Diet and Hydration**")
            print(" - Avoid heavy meals, caffeine, and alcohol in the hours before bed.")
            print(" - Eat a balanced diet with magnesium-rich foods (e.g., leafy greens, nuts) to support relaxation.")
            print(" - Try warm milk or herbal teas, such as lavender or lemon balm, which are known to support better sleep.")

            print("\n**Physical Activity**")
            print(" - Aim for regular exercise during the day, as it can help reduce stress and promote restful sleep.")
            print(" - Avoid vigorous exercise within two hours of bedtime, as it can increase energy and make it harder to fall asleep.")

            print("\n**Sleep Environment**")
            print(" - Keep your bedroom cool, dark, and quiet.")
            print(" - Use earplugs, an eye mask, or a white noise machine to block out disturbances.")
            print(" - Limit exposure to screens before bed; the blue light from devices can interfere with melatonin production.")

            print("\n**Natural Remedies**")
            print(" - Try drinking valerian root tea or chamomile tea before bed for relaxation.")
            print(" - Lavender essential oil can be used in a diffuser to create a calming environment.")
            print(" - Epsom salt baths before bed can help relax muscles and reduce stress.")

        elif self.sleep_quality == "Fair":
            print("\nAdvice: Your sleep quality is okay, but there are still areas that could be improved.")

            print("\n**Sleep Routine**")
            print(" - Try to maintain a consistent sleep schedule, even on weekends.")
            print(" - Wind down before bed with a calming pre-sleep routine, such as reading a book or listening to soft music.")

            print("\n**Stress Reduction**")
            print(" - Introduce calming activities like gentle yoga or breathing exercises before bed.")
            print(" - Try natural relaxants, like chamomile tea or valerian supplements, to help ease tension.")

            print("\n**Diet and Exercise**")
            print(" - Keep a balanced diet rich in fruits, vegetables, and whole grains.")
            print(" - Moderate caffeine intake, especially in the afternoon, and stay hydrated throughout the day.")
            print(" - Engage in regular physical activity to help manage stress and improve sleep quality.")

            print("\n**Sleep Environment**")
            print(" - Make sure your bedroom is quiet, cool, and comfortable.")
            print(" - Consider using essential oils, like lavender or eucalyptus, to create a soothing atmosphere for sleep.")
            print(" - Reduce screen time before bed, or use blue light filters to minimize the impact on your sleep cycle.")

            print("\n**Natural Remedies**")
            print(" - Drink chamomile tea or a mixture of lemon balm and honey before bedtime to help you relax.")
            print(" - Use a diffuser with lavender or cedarwood essential oil to promote relaxation.")
            print(" - A warm bath with a few drops of eucalyptus or lavender essential oil can help ease your mind before sleep.")

        elif self.sleep_quality == "Good":
            print("\nAdvice: You’re doing well! A few tweaks could further enhance your sleep quality.")

            print("\n**Consistency**")
            print(" - Continue going to bed and waking up at similar times daily to reinforce your sleep cycle.")
            print(" - Avoid changes in your schedule over weekends, as this can disrupt your internal clock.")

            print("\n**Relaxation Techniques**")
            print(" - Try progressive muscle relaxation to release any lingering physical tension before bed.")
            print(" - Practice mindful breathing techniques to promote a calm mind and smooth transition into sleep.")

            print("\n**Diet and Exercise**")
            print(" - Maintain a balanced diet and avoid eating large meals too close to bedtime.")
            print(" - Exercise regularly, but be cautious with evening workouts as they can increase alertness.")

            print("\n**Natural Remedies**")
            print(" - A warm cup of chamomile or valerian tea before bed can promote restful sleep.")
            print(" - Try a warm bath with a handful of Epsom salts and lavender essential oil.")
            print(" - A few drops of lavender or frankincense essential oils on your pillow can help you relax.")

            print("\n**Sleep Environment**")
            print(" - Ensure your bedroom is cool, dark, and quiet, as these conditions are optimal for sleep.")
            print(" - Use blackout curtains or a sleep mask if light is an issue, and a white noise machine if noise is disruptive.")
        else:
            print("\nAdvice: Fantastic! You’re maintaining excellent sleep hygiene.")

            print("\n**Routine Consistency**")
            print(" - Keep up with your regular sleep schedule and habits.")
            print(" - Continue to avoid screens and stimulating activities close to bedtime.")

            print("\n**Stress Management**")
            print(" - Keep stress levels low with consistent relaxation practices like deep breathing or gentle yoga.")

            print("\n**Physical Activity**")
            print(" - Continue regular exercise and consider integrating mindfulness or meditation into your day.")

            print("\n**Diet**")
            print(" - Maintain a healthy diet with nutrient-rich foods that support overall well-being.")
            print(" - Try foods like almonds, turkey, or kiwi, which have properties that can support better sleep quality.")

            print("\n**Environment and Natural Remedies**")
            print(" - Keep your sleep environment serene and relaxing.")
            print(" - If you ever need extra relaxation, try soothing herbs like chamomile or lavender essential oil.")
            print(" - A light stretching routine or warm bath can be relaxing rituals to prepare you for deep, restful sleep.")

            print("\n**Few more Natural Remedies**")
            print(" - Chamomile tea, valerian root, or a few drops of lavender essential oil can enhance your evening routine.")
            print(" - Consider a warm bath with Epsom salts and essential oils like lavender or frankincense.")
            print(" - Incorporate stretching or yoga in the evening to reduce tension and promote sleep.")

    def plot_and_save_individual_graphs(self):
      factors = [
        ("Sleep Hours", self.sleep_hours, 6, 8, "Hours"),
        ("Stress Level", self.stress_level, 1, 3, "Level"),
        ("Exercise Level", self.exercise_level, 2, 5, "Days"),
        ("Diet Quality", self.diet_quality, 8, 10, "Rating"),
        ("Sleep Schedule Consistency", self.sleep_schedule_consistency, 5, 7, "Days/Week"),
        ("Caffeine Intake", self.caffeine_intake, 0, 2, "Cups"),
        ("Screen Time", self.screen_time, 0, 2, "Hours"),
        ("Work Stress", self.stress_level, 1, 3, "Level"),

      ]
      graph_files = []
      for factor_name, user_value, ideal_min, ideal_max, unit in factors:
        print(f"Generating graph for {factor_name}...")  # Debugging output
        file_name = self.plot_and_save_factor(factor_name, user_value, ideal_min, ideal_max, unit)
        print(f"Graph saved as {file_name}")  # Debugging output
        graph_files.append(file_name)
      return graph_files



    def plot_and_save_factor(self, factor_name, user_value, ideal_min, ideal_max, unit):
        # Generate and save individual factor graph
        plt.figure(figsize=(8, 6))

        # Plot user value vs ideal range
        plt.bar(factor_name, user_value, color='skyblue', label='Your Value')
        plt.axhline(y=ideal_min, color='green', linestyle='--', label=f'Ideal Min: {ideal_min}')
        plt.axhline(y=ideal_max, color='green', linestyle='--', label=f'Ideal Max: {ideal_max}')

        plt.xlabel(factor_name)
        plt.ylabel(unit)
        plt.title(f'{factor_name} vs Ideal Range')
        plt.legend()
        plt.tight_layout()

        # Save graph to a file
        file_name = f"{factor_name}_graph.png"
        plt.savefig(file_name)
        plt.close()

        return file_name

    def add_page_border(self, canvas_obj):
        """Adds a border to the current page."""
        canvas_obj.rect(50, 50, letter[0] - 100, letter[1] - 100, stroke=1, fill=0) # Adjust margins as needed

    def generate_pdf_report(self):
     file_name = "Sleep_Quality_Report.pdf"
     c = canvas.Canvas(file_name, pagesize=letter)
     c.setFont("Helvetica", 12)

     # Add border to the first page
     self.add_page_border(c)

     # Add sleep quality and advice
     c.drawString(100, 750, "Sleep Quality Report")
     c.drawString(100, 730, f"Overall Sleep Quality: {self.sleep_quality}")
     y_position = 710

     # Write advice text dynamically with page breaks
     y_position = self.write_advice(self.sleep_quality, c,y_position)

     # Add graphs on new pages
     graph_files = self.plot_and_save_individual_graphs()

     # Check if graphs are being generated correctly
     for graph_file in graph_files:
        print(f"Adding {graph_file} to PDF...")
        if os.path.exists(graph_file):  # Check if the file exists before adding to PDF
            c.showPage()  # New page for each graph
            c.drawImage(graph_file, 50, 300, width=500, height=400)
        else:
            print(f"Graph file {graph_file} not found.")

     # Save the PDF
     c.save()
     print(f"PDF report generated: {file_name}")

     # Clean up the saved graph images
     for graph_file in graph_files:
        if os.path.exists(graph_file):
            os.remove(graph_file)  # Remove the graph file after adding to PDF
        else:
            print(f"File not found: {graph_file}")


    def write_advice(self, sleep_quality, c,y_position):
        advice_text = {
            "Poor": [
                "Advice: You may need to make several lifestyle adjustments to improve your sleep quality.",
                "**Sleep Duration and Routine**", "- Aim for at least 7-8 hours of sleep each night.", " - Set a regular bedtime and wake-up time, even on weekends."," - Avoid long naps during the day, as they can disrupt nighttime sleep.",
                "**Stress and Relaxation**", "- Try relaxation techniques like deep breathing,progressive muscle relaxation, or mindfulness meditation."," - Consider natural remedies, such as drinking chamomile or valerian tea before bed to help calm your mind."," - Journaling before bed can help clear your mind of stressful thoughts.",
                "**Diet and Hydration**", "- Avoid heavy meals, caffeine, and alcohol before bed.", " - Eat a balanced diet with magnesium-rich foods (e.g., leafy greens, nuts) to support relaxation.", " - Try warm milk or herbal teas, such as lavender or lemon balm, which are known to support better sleep.",
                "**Physical Activity**", "- Aim for regular exercise during the day,as it can help reduce stress and promote restful sleep.", " - Avoid vigorous exercise within two hours of bedtime, as it can increase energy and make it harder to fall asleep.",
                "**Natural Remedies**", " - Try drinking valerian root tea or chamomile tea before bed for relaxation."," - Lavender essential oil can be used in a diffuser to create a calming environment."," - Epsom salt baths before bed can help relax muscles and reduce stress."
            ],
            "Fair": [
                "Advice: Your sleep quality is okay, but there are still areas that could be improved.",
                "**Sleep Routine**", "- Try to maintain a consistent sleep schedule,even on weekends."," - Wind down before bed with a calming pre-sleep routine, such as reading a book or listening to soft music.",
                "**Stress Reduction**", "- Introduce calming activities like gentle yoga or breathing exercises before bed.", " - Try natural relaxants, like chamomile tea or valerian supplements, to help ease tension.",
                "**Diet and Exercise**", "- Keep a balanced diet rich in fruits, vegetables, and whole grains.", " - Moderate caffeine intake, especially in the afternoon, and stay hydrated throughout the day.", " - Engage in regular physical activity to help manage stress and improve sleep quality.",
                "**Sleep Environment**", "- Make sure your bedroom is quiet, cool, and comfortable.", " - Consider using essential oils, like lavender or eucalyptus, to create a soothing atmosphere for sleep.", " - Reduce screen time before bed, or use blue light filters to minimize the impact on your sleep cycle.",
                "**Natural Remedies**", "- Drink chamomile tea or a mixture of lemon balm and honey before bedtime to help you relax.", " - Use a diffuser with lavender or cedarwood essential oil to promote relaxation."," - A warm bath with a few drops of eucalyptus or lavender essential oil can help ease your mind before sleep."
            ],
            "Good": [
                "Advice: Your sleep quality is quite good, but you can always improve.",
                "**Consistency**", "- Continue going to bed and waking up at similar times daily to reinforce your sleep cycle.", " - Avoid changes in your schedule over weekends, as this can disrupt your internal clock.",
                "**Relaxation Techniques**", "- Try progressive muscle relaxation to release any lingering physical tension before bed.", " - Practice mindful breathing techniques to promote a calm mind and smooth transition into sleep.",
                "**Diet and Exercise**", "- Maintain a balanced diet and avoid eating large meals too close to bedtime.", " - Exercise regularly, but be cautious with evening workouts as they can increase alertness.",
                "**Natural Remedies**", "- A warm cup of chamomile or valerian tea before bed can promote restful sleep.", " - Try a warm bath with a handful ofbackends and essential oils like lavender or frankincense."," - A few drops of lavender or frankincense essential oils on your pillow can help you relax.",
                "**Sleep Environment**", "- Ensure your bedroom is cool, dark, and quiet, as these conditions are optimal for sleep.", " - Use blackout curtains or a sleep mask if light is an issue, and a white noise machine if noise is disruptive."

            ],
            "Excellent": [
                "Advice: Fantastic! You’re maintaining excellent sleep hygiene.",
                "**Routine Consistency**"," - Keep up with your regular sleep schedule and habits."," - Continue to avoid screens and stimulating activities close to bedtime.",
                "**Stress Management**", "- Keep stress levels low with consistent relaxation practices like deep breathing or gentle yoga.",
                "**Physical Activity**", "- Continue regular exercise and consider integrating mindfulness or meditation into your day.",
                "**Diet**", "- Maintain a healthy diet with nutrient-rich foods that support overall well-being."," - Try foods like almonds, turkey, or kiwi, which have properties that can support better sleep quality.",
                "**Environment and Natural Remedies**", "- Keep your sleep environment serene and relaxing.", " - If you ever need extra relaxation, try soothing herbs like chamomile or lavender essential oil.", " - A light stretching routine or warm bath can be relaxing rituals to prepare you for deep, restful sleep."
            ]
        }
        new_y_position = y_position  # Initialize new_y_position

        for line in advice_text[sleep_quality]:
            if new_y_position < 50:
                c.showPage()
                new_y_position = 750

            # Split long lines into shorter lines
            line_width = c.stringWidth(line, "Helvetica", 12)  # Get width of line
            page_width = letter[0] - 200  # Usable page width (considering margins)

            if line_width > page_width:  # If line is too wide
                words = line.split()
                current_line = ""
                for word in words:
                    if c.stringWidth(current_line + word + " ", "Helvetica", 12) <= page_width:
                        current_line += word + " "
                    else:
                        c.drawString(100, new_y_position, current_line.strip())  # Draw current line
                        new_y_position -= 20  # Move to next line
                        if new_y_position < 50:  # Check for page break
                            c.showPage()
                            new_y_position = 750
                        current_line = word + " "  # Start new line with current word
                c.drawString(100, new_y_position, current_line.strip())  # Draw remaining part of the line
            else:  # If line fits within page width
                c.drawString(100, new_y_position, line)  # Draw the line as it is

            new_y_position -= 20

        return new_y_position

    def display_main_menu(self):
     print("\nWelcome to the Sleep Quality Advisor!")
     print("1. I want to see only advice")
     print("2. I want to analyze my sleep")
     print("3. Both: See advice and analyze sleep")
     print("4. Download PDF Report")
     choice = int(input("Enter the number corresponding to your choice: "))

     if choice == 1:
        self.provide_advice()  # Show only advice
     elif choice == 2:
        # If option 2 is selected, generate and display graphs
        print("\nGenerating sleep quality graphs...")
        graph_files = self.plot_and_save_individual_graphs()

        # Display the graphs (e.g., open them in the notebook or prepare for download)
        for graph_file in graph_files:
            print(f"Displaying graph: {graph_file}")
            # If running in a Jupyter notebook, you can display the graph inline using Image from IPython.display
            from IPython.display import Image
            display(Image(filename=graph_file))
     elif choice == 3:
        self.provide_advice()  # Provide advice based on the evaluation
        print("\nGenerating sleep quality graphs...")
        graph_files = self.plot_and_save_individual_graphs()

        # Display the graphs (e.g., open them in the notebook or prepare for download)
        for graph_file in graph_files:
            print(f"Displaying graph: {graph_file}")
            # If running in a Jupyter notebook, you can display the graph inline using Image from IPython.display
            from IPython.display import Image
            display(Image(filename=graph_file))
     elif choice == 4:
        self.generate_pdf_report()  # Generate PDF report
     else:
        print("Invalid choice. Please select a valid option (1,2,3, or 4).")
        self.display_main_menu()


# Usage
advisor = SleepQualityAdvisor()
advisor.ask_questions()
advisor.evaluate_sleep_quality()
advisor.display_main_menu()


files.download("Sleep_Quality_Report.pdf")
```
